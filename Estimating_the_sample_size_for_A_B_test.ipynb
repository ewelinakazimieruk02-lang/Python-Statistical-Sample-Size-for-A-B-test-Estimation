```python
# Część 1
# Oczekiwany efekt w postaci 3% różnicy w konwersji.
# Poziom istotności to 5%, a moc testu 80%.
# Obecny współczynnik konwersji wynosi 10%.

import numpy as np
import statsmodels.stats.api as sms
from statsmodels.stats.proportion import proportion_effectsize

# Baseline conversion rate (10%)
baseline_conversion = 0.10

# Expected effect (3% relative difference in conversion)
expected_lift = 0.03 * baseline_conversion   # 3% * 10% = 0.003
variant_conversion = baseline_conversion + expected_lift  # 0.10 + 0.003 = 0.103 (10.3%)

# Effect size calculation (Cohen's h)
effect_size = proportion_effectsize(baseline_conversion, variant_conversion)

# Test parameters
alpha = 0.05
power = 0.80

# Sample size calculation
sample_size = sms.NormalIndPower().solve_power(
    effect_size=effect_size,
    power=power,
    alpha=alpha,
    ratio=1
)

print(f"Required sample size for each group: {int(np.ceil(sample_size))}")
```

Dla wzrostu konwersji z 10% do około 10.3% (czyli o 3% względnie), przy poziomie istotności 0.05 i mocy testu 80%, potrzebujemy około 159 059 użytkowników w każdej grupie. Tak duża liczba wynika z tego, że oczekiwany efekt jest bardzo mały, więc potrzeba ogromnej próby, aby móc go wiarygodnie wykryć. W praktyce oznacza to, że test będzie bardzo kosztowny i czasochłonny.

```python
# Część 2
#Jak zmieni się wymagana wielkość próby, jeśli obniżysz poziom istotności do 1%.
# zmieniamy tylko poziom istotności

import numpy as np
import statsmodels.stats.api as sms
from statsmodels.stats.proportion import proportion_effectsize

baseline_conversion = 0.10
expected_lift = 0.03 * baseline_conversion
variant_conversion = baseline_conversion + expected_lift

effect_size = proportion_effectsize(baseline_conversion, variant_conversion)

alpha = 0.01   # 1%
power = 0.80

sample_size_alpha_1 = sms.NormalIndPower().solve_power(
    effect_size=effect_size,
    power=power,
    alpha=alpha,
    ratio=1
)

print(f"Required sample size for each group (alpha=0.01): {int(np.ceil(sample_size_alpha_1))}")

```

Po obniżeniu poziomu istotności z 0.05 do 0.01, przy tym samym oczekiwanym wzroście konwersji z 10% do około 10.3%, wymagana liczba użytkowników w każdej grupie rośnie z około 159 059 do około 236 676. Oznacza to, że test staje się jeszcze bardziej rygorystyczny i potrzebuje znacznie więcej danych, aby uznać różnicę za istotną statystycznie. W praktyce taki eksperyment będzie trwał jeszcze dłużej i będzie jeszcze droższy niż w przypadku alpha = 0.05.

```python
# Część 3
# oczekiwany efekt zmniejszy się do 1% różnicy w konwersji, jaka będzie wymagana wielkość próby?

import numpy as np
import statsmodels.stats.api as sms
from statsmodels.stats.proportion import proportion_effectsize

# Baseline conversion rate (10%)
baseline_conversion = 0.10

# Expected effect (1% relative difference in conversion)
expected_lift_small = 0.01 * baseline_conversion   # 1% * 10% = 0.001
variant_conversion_small = baseline_conversion + expected_lift_small  # 0.10 + 0.001 = 0.101 (10.1%)

# Effect size calculation (Cohen's h)
effect_size_small = proportion_effectsize(baseline_conversion, variant_conversion_small)

# Test parameters
alpha = 0.05
power = 0.80

# Sample size calculation for smaller effect
sample_size_small_effect = sms.NormalIndPower().solve_power(
    effect_size=effect_size_small,
    power=power,
    alpha=alpha,
    ratio=1
)

print(f"Required sample size for each group (1% relative difference): {int(np.ceil(sample_size_small_effect))}")

```

Dla bardzo małego efektu, czyli wzrostu konwersji z 10% do około 10.1% (czyli o 1% względnie), potrzebujemy aż około 1 419 062 użytkowników w każdej grupie. Tak ogromna liczba wynika z tego, że różnica między wersjami jest minimalna, więc potrzeba bardzo dużej próby, aby móc ją statystycznie wykryć. W porównaniu z efektem 3% względnie (gdzie potrzeba było około 159 059 osób), widać wyraźnie, że im mniejszy oczekiwany efekt, tym test staje się znacznie bardziej wymagający, kosztowny i czasochłonny.
